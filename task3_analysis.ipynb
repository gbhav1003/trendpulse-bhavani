```python
# task3_analysis.py

# Import required libraries
import pandas as pd
import numpy as np
import os

# -------------------------------
# STEP 1: Load cleaned CSV file
# -------------------------------
file_path = "data/trends_clean.csv"

try:
    df = pd.read_csv(file_path)
    print(f"Loaded {len(df)} rows from {file_path}\n")
except Exception as e:
    print("Error loading cleaned CSV:", e)
    exit()

# Print first 5 rows
print("First 5 rows of dataset:")
print(df.head())

# Print shape of dataframe
print("\nShape of DataFrame:", df.shape)

# Print average score and average comments
avg_score = df["score"].mean()
avg_comments = df["num_comments"].mean()

print("\nAverage score:", round(avg_score, 2))
print("Average num_comments:", round(avg_comments, 2))

# -------------------------------
# STEP 2: Analysis using NumPy
# -------------------------------

score_array = np.array(df["score"])

print("\nNumPy Statistical Analysis on Score:")
print("Mean score:", round(np.mean(score_array), 2))
print("Median score:", np.median(score_array))
print("Standard deviation of score:", round(np.std(score_array), 2))
print("Highest score:", np.max(score_array))
print("Lowest score:", np.min(score_array))

# Which category has most stories
top_category = df["category"].value_counts().idxmax()
print("\nCategory with most stories:", top_category)

# Story with most comments
max_comment_index = df["num_comments"].idxmax()
top_commented_story = df.loc[max_comment_index]

print("\nStory with highest comments:")
print("Title:", top_commented_story["title"])
print("Comment count:", top_commented_story["num_comments"])

# -------------------------------
# STEP 3: Add new columns
# -------------------------------

# engagement = num_comments / (score + 1)
df["engagement"] = df["num_comments"] / (df["score"] + 1)

# is_popular = True if score > average score else False
df["is_popular"] = df["score"] > avg_score

print("\nNew columns added successfully: engagement, is_popular")

# -------------------------------
# STEP 4: Save analysed file
# -------------------------------

output_file = "data/trends_analysed.csv"
df.to_csv(output_file, index=False)

print(f"\nSaved analysed dataset to {output_file}")
```
